```
import pandas as pd

df = pd.read_csv('heart.csv')
df = df.sample(frac=1)
print(df)
```

```
X_train = df.iloc[:273, :-1]
X_train = X_train.to_numpy()
print(X_train.shape)

X_test = df.iloc[273:, :-1]
X_test = X_test.to_numpy()
print(X_test.shape)
```

```
Y_train = df.iloc[:273, -1:]
Y_train = Y_train.to_numpy()
print(Y_train.shape)

Y_test = df.iloc[273:, -1:]
Y_test = Y_test.to_numpy()
print(Y_test.shape)
```

```
model = keras.Sequential()

model.add(keras.layers.Flatten(input_shape=(13,)))
model.add(keras.layers.Dense(16, activation=keras.activations.relu))
model.add(keras.layers.Dense(1, activation=keras.activations.sigmoid))

model.summary()

model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.binary_crossentropy, metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=300)
```

```
test_loss, test_acc = model.evaluate(X_test, Y_test)

print(test_acc)
```
